```python
import networkx as nx
import graphviz
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
G = nx.complete_graph(3)
pos = {0: (0, 2), 1: (2, 2), 2: (1, 1)}
# nodes
options = {"edgecolors": "tab:gray", "node_size": 3000, "alpha": 0.9}
nx.draw_networkx_nodes(G, pos, nodelist=[0, 1, 2], node_color="black", **options)
# edges
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.0)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[(0,1), (1, 2), (2, 0)],
    width=8,
    alpha=0.7,
    edge_color="black",
)

# some math labels
labels = {}
labels[0] = r"$1$"
labels[1] = r"$2$"
labels[2] = r"$3$"
nx.draw_networkx_labels(G, pos, labels, font_size=18, font_color="whitesmoke")

plt.axis("off")
plt.show()
```
